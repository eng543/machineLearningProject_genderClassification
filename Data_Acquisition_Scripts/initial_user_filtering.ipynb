```python
import pandas as pd
import ast
```

```python
#filename = 'results10.txt'
filename = 'results11.txt'
#filename = 'results12.txt'
```

```python
dict_form = []
```

```python
with open(filename, 'U') as rf:
    str_form = rf.read().splitlines()
    for i in range(0, len(str_form)):
        dict_form.append(ast.literal_eval(str_form[i]))
```

```python
users = pd.DataFrame()

for i in range(0, len(dict_form)):
    temp = pd.DataFrame.from_dict(dict_form[i], orient = 'index')
    users = users.append(temp)

users['user_id'] = users.index
```

```python
cols = list(users)
cols.insert(0, cols.pop(cols.index('user_id')))
users = users.ix[:, cols]
users.columns = ['user_id', 'q_keyword', 'location', 'language', 'description']
users = users.reset_index(drop = True)
```

```python
print users[0:5], len(users)
```

```python
# get rid of duplicates
users_dupes = users.drop_duplicates()
```

```python
# remove users whose language != "en"
users_lang = users_dupes[users_dupes['language'].isin(['en'])]
users_lang = users_lang.reset_index(drop = True)
```

```python
# users who do not use keyword in description
delete = []
for i in range(0, len(users_lang)):
    if users_lang['q_keyword'].ix[i] not in users_lang['description'].ix[i].lower():
        delete.append(i)

users_keyword = users_lang.drop(users_lang.index[delete])
```
