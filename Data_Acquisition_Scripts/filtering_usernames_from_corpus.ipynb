```python
import pandas as pd
```

```python
tweets = pd.read_csv("tweets.csv")
```

```python
tweets
```

```python
# remove users whose language != "en"
tweets_lang = tweets[tweets['Language'].isin(['en'])]
tweets_lang = tweets_lang.reset_index(drop = True)
```

```python
tweets_lang
```

```python
usernames = {}
```

```python
keywords = ['mom', 'dad', 'mother', 'father', 'sister', 'brother',
           'aunt', 'uncle', 'gramma', 'grampa', 'grandma', 'grandpa',
           'mama', 'momma', 'papa', 'male', 'female', 'man', 'woman',
           'girl', 'boy', 'trans', 'queer', 'gender', 'binary', 'fluid',
           'andro', 'eunuch', 'MTF', 'FTM', 'MTX', 'FTX']

for i in range(0, len(tweets_lang)):
    for keyword in keywords:
        try:
            if keyword in tweets_lang['Bio'].ix[i].lower():
                if tweets_lang['Nickname'].ix[i] in usernames:
                    usernames[tweets_lang['Nickname'].ix[i]].append(keyword)
                else:
                    usernames[tweets_lang['Nickname'].ix[i]] = [keyword]
        except:
            continue
```

```python
len(usernames)
```

```python
usernames
```

```python
# filter usernames to get unique values
# subset usernames
# one set with a single value
# another set with two values
singleUsernames = {}
multipleUsernames = {}

for key in usernames.keys():
    usernames[key] = list(set(usernames[key]))
    if len(usernames[key]) == 1:
        singleUsernames[key] = usernames[key]
    else:
        multipleUsernames[key] = usernames[key]
```

```python
# write table with usernames for get_tweet script
output = open('usernames.txt', 'w')

count = 1
for key in singleUsernames.keys():
    output.write(str(count))
    output.write("\t")
    output.write(str(key))
    output.write("\t")
    output.write("")
    output.write("\n")
    count += 1
    
output.close()
```

```python
output = open('username_genders.txt', 'w')

for key in singleUsernames.keys():
    output.write(str(key))
    output.write("\t")
    output.write(singleUsernames[key][0])
    output.write("\n")
    
output.close()
```
